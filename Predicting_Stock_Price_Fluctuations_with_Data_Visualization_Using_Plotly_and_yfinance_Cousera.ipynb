```python
import yfinance as yf
import pandas as pd

# Mengambil data saham Tesla dari Yahoo Finance
tesla_data = yf.download('TSLA', start='2010-01-01', end='2024-01-01')

# Mereset index dan menampilkan lima baris pertama
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

```python
import requests
from bs4 import BeautifulSoup

# URL untuk pendapatan Tesla dari Yahoo Finance
url = "https://finance.yahoo.com/quote/TSLA/financials?p=TSLA"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Menemukan tabel yang berisi pendapatan (sesuaikan dengan struktur halaman)
table = soup.find_all('table')[0]  # Ambil tabel pertama
rows = table.find_all('tr')

# Mengambil data pendapatan
revenue_data = []
for row in rows[1:]:  # Abaikan header
    cols = row.find_all('td')
    if len(cols) > 1:
        year = cols[0].text.strip()
        revenue = cols[1].text.strip()
        revenue_data.append([year, revenue])

# Membuat DataFrame dari data yang ditemukan
tesla_revenue = pd.DataFrame(revenue_data, columns=['Year', 'Revenue'])
tesla_revenue.tail()
```

```python
import yfinance as yf
import pandas as pd
import plotly.express as px

# Mengambil data saham Tesla dari Yahoo Finance
tesla_data = yf.download('TSLA', start='2010-01-01', end='2024-01-01')

# Mereset index dan memastikan kolom 'Date' dalam format datetime
tesla_data.reset_index(inplace=True)

# Menampilkan nama kolom untuk memverifikasi struktur DataFrame
print("Kolom sebelum modifikasi:", tesla_data.columns)

# Memastikan kolom-kolom yang diinginkan
tesla_data.columns = tesla_data.columns.droplevel(1)

# Menampilkan nama kolom setelah modifikasi
print("Kolom setelah modifikasi:", tesla_data.columns)

# Menghapus baris yang memiliki nilai kosong di kolom 'Close' (jika ada)
if 'Close' in tesla_data.columns:
    tesla_data = tesla_data.dropna(subset=['Close'])
else:
    print("Kolom 'Close' tidak ditemukan!")

# Fungsi untuk membuat grafik saham
def make_graph(df, stock_name):
    fig = px.line(df, x='Date', y='Close', title=f'{stock_name} Stock Price Over Time')
    fig.show()

# Membuat grafik saham Tesla
make_graph(tesla_data, 'Tesla')
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Mengunduh data saham GameStop (GME) menggunakan yfinance
gme_data = yf.download('GME', start='2010-01-01', end='2023-12-31')

# Membuat grafik menggunakan Pandas plot
gme_data['Close'].plot(title="Harga Saham GameStop", figsize=(10, 6))
plt.xlabel('Tanggal')
plt.ylabel('Harga Penutupan')
plt.show()
```
